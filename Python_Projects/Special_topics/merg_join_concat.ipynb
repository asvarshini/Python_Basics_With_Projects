```python
#Merge (Combine tables based on a common column)
import pandas as pd
df = pd.DataFrame([[1,"ram"],[2,"sam"],[3,"var"]], columns=["id","name"])
print(df.to_string(index=False))
```

```python
df2={
    'costmer_id':[101,102,103,104],
    "id":[1,2,4,5],
    "amount":[10000,2000,3000,4000]
    }
print(pd.DataFrame(df2).to_string(index=False))
```

```python
#Basic Merge Syntax
df2_df = pd.DataFrame(df2) # Convert the dictionary df2 to a DataFrame
pd.merge(df, df2_df, on="id", how="inner") # Merge on the common 'id' column
```

```python
# 1 Inner merge
pd.merge(df, df2_df, on="id", how="inner") # Merge on the common 'id' column
```

```python
# 2 Left Merge
pd.merge(df, df2_df, on="id", how="left")
```

```python
#RightMerge
pd.merge(df, df2_df, on="id", how="right")
```

```python
#outerMerge
pd.merge(df, df2_df, on="id", how="outer")
```

```python
#2. JOIN "Merge using index instead of column"
df.join(df2_df.set_index('id'),on='id',lsuffix='_left',rsuffix='_right')
```

```python
#3. CONCAT "Just attach data"
pd.concat([df,df2_df],axis=1)

```

```python
import pandas as pd

customers = pd.DataFrame({
    "id": [1,2,3],
    "name": ["A", "B", "C"]
})

orders = pd.DataFrame({
    "id": [1,2,4],
    "amount": [100,200,300]
})

# Try all joins
print(pd.merge(customers, orders, on="id", how="inner"))
print(pd.merge(customers, orders, on="id", how="left"))
print(pd.concat([customers, orders], axis=1))
```

*map function means applying funtion to every iteam in the collection


*syntax =map(function,iterable) functtion(funcyion=what di u want ,iterable=[list,tuple])

```python
#Normal function
lis=[3,5,6,8]
for eachnumber in lis:
  print(eachnumber *2)
```

```python
#by using map and lambda
l=[3, 5, 6, 8]
result = list(map(lambda x: x * 2,l))
print(result)
```

#Fliter
* Syntax filter(function,iterable)

```python
#using filter normal without lambda
def is_even(x):
  if x%2==0:
    return x
num=[1,2,4,5]
print(list(filter(is_even,num)))
```

```python
#using lambda and list
num=[9,8,7,6]
is_even=list(filter(lambda x:x%2==0, num))
print(is_even)
```

```python
names = ["Ram", "John", "Sam", "Alex"]

result = list(filter(lambda x: len(x)> 3, names))

print(result)
```

#sort with lambda
somtimes you have to sort
length of words
second value in tuple
custom logic
*******syntax*********
list.sort(key = lambda x: condition)

```python
#Sort by Last Digit
list=[1,5,7,9]
list.sort(key=lambda x:x%10)
print(list)
```

```python
#Sort Words by Length
fruits=["papaya",'promaganate','grapes','kiwi']
fruits.sort(key=lambda x:len(x))
print(fruits)
```

```python
#Sort Words by Length
students = [("Ram", 85), ("John", 90), ("Sam", 75)]
students.sort(key=lambda x:x[1])
print(students)
```
